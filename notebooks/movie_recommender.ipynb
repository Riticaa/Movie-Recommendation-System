```python
import ast

def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])

        return l

```

```python
def convert_cast(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i['name'])
        counter +=1
            
        

    return l
```

```python
movies['cast']= movies['cast'].apply(convert_cast)
movies.head(2)
```

```python
def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
            break
            
        

    return l    
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
def join_tags(x):
    # Check if x is iterable (like a list) and not a string
    if isinstance(x, (list, tuple)):
        return " ".join(x)
    # If x is already a string, return it as is
    elif isinstance(x, str):
        return x
    # If x is None or another type, return an empty string or handle as needed
    else:
        return ""

# Apply the custom function instead of directly using join
new_df['tags'] = new_df['tags'].apply(join_tags)
```

```python
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
new_df['tags']= new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
new_df.shape
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(l)   
```

```python
new_df['tags']= new_df['tags'].apply(stems)
```

```python
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer( max_features= 5000, stop_words= 'english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_df[new_df['title'] =='Spider-Man'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True, key = lambda x: x[1])
    for i in distance[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
